```python
import fitz  # PyMuPDF library for handling PDF documents
import re  # Regular expressions library for pattern matching
import csv  # CSV library for writing data to CSV files

# Open the PDF document
doc = fitz.open("C:/Users/user/Desktop/pdfs/120_24pgs_mlinterviewquestions.pdf")
text = ""  # Initialize an empty string to store the extracted text

# Iterate over each page in the PDF and extract text
for page in doc:
    text += page.get_text()

# Initialize an empty list to store question-answer pairs
question_answer_pairs = []

# Define a regular expression pattern to match questions and answers
pattern = r'(\d+\.\s+.*?)(?=\s*Answer:|Answers:|(?:\d+\.\s))'

# Find all matches in the extracted text using the defined pattern
matches = re.findall(pattern, text, re.DOTALL)

# Process each match to separate questions and answers
for match in matches:
    if "Answer:" in match:
        # Split the match into question and answer using "Answer:" as the delimiter
        question, answer = match.split("Answer:", 1)
    elif "Answers:" in match:
        # Split the match into question and answer using "Answers:" as the delimiter
        question, answer = match.split("Answers:", 1)
    else:
        # If no answer section is found, treat the entire match as the question
        question = match.strip()
        answer = ""

    # Remove any leading or trailing whitespace from the question and answer
    question = question.strip()
    answer = answer.strip()

    # Append the processed question-answer pair to the list
    question_answer_pairs.append([question, answer])

# Specify the output CSV file path
csv_file = "C:/Users/user/Desktop/pdfs/questions_and_answers.csv"

# Write the question-answer pairs to a CSV file
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Question', 'Answer'])  # Write the header row
    writer.writerows(question_answer_pairs)  # Write the question-answer data

# Print a message indicating successful creation of the CSV file
print(f"CSV file has been created at {csv_file}")
```

```python
import fitz  # PyMuPDF library for handling PDF documents
import pandas as pd  # Pandas library for data manipulation and storage
import re  # Regular expressions library for pattern matching

# Open the PDF document
doc = fitz.open("C:/Users/user/Desktop/pdfs/120_24pgs_mlinterviewquestions.pdf")
text = ""  # Initialize an empty string to store the extracted text

# Extract text from each page in the PDF and concatenate it
for page in doc:
    text += page.get_text()

# Initialize empty lists to store questions and answers separately
questions = []
answers = []

# Define a regex pattern to extract questions and answers
# The pattern matches:
# 1. A number followed by a period (question number)
# 2. The question text (non-greedy)
# 3. "Answer" or "Answers" followed by the answer text
# Uses DOTALL to include newline characters in matching.
pattern = re.compile(r'(\d+)\.\s*(.*?)\s*Answers?:\s*(.*?)(?=\d+\.\s|\Z)', re.DOTALL)

# Find all matches based on the regex pattern
matches = pattern.findall(text)

# Process each match to separate questions and answers
for match in matches:
    question_number, question_text, answer_text = match
    questions.append(question_text.strip())  # Clean and add the question text
    answers.append(answer_text.strip())  # Clean and add the answer text

# Create a DataFrame to store questions and answers in a structured format
qa_df = pd.DataFrame({
    "Question": questions,
    "Answer": answers
})

# Save the DataFrame to a CSV file at the specified location
qa_df.to_csv("C:/Users/user/Desktop/pdfs/ml_interview_questions.csv", index=False, encoding="utf-8")

# Print a confirmation message after saving the CSV file
print("File saved as ml_interview_questions.csv")
```
