```python
import csv
def append_in_csv(text):
    try:
        with open('wikipedia.csv', mode='r', newline='', encoding='utf-8') as file:
            reader = csv.reader(file)
            existing_rows = [row[0] for row in reader]  # Assuming each row contains one column (text)
            
            # Check if the text is already in the CSV
            if text not in existing_rows:
                with open('wikipedia.csv', mode='a', newline='', encoding='utf-8') as file_append:
                    writer = csv.writer(file_append)
                    writer.writerow([text])  # Append the new text
            else:
                pass
    except FileNotFoundError:
        # If the CSV file doesn't exist yet, create it and write the first row
        with open('wikipedia.csv', mode='a', newline='', encoding='utf-8') as file_append:
            writer = csv.writer(file_append)
            writer.writerow([text])  # Write the text as a new row
        print(f"Created file and added first row: {text}")
```

```python
import wikipedia

def fetch_wikipedia_summary(topic, sentences=20):
    try:
        # Fetch summary with the specified number of sentences
        summary = wikipedia.summary(topic, sentences=sentences)
        return summary
    except wikipedia.exceptions.DisambiguationError as e:
        # If the topic is ambiguous, suggest possible options
        print(f"Disambiguation error: The topic '{topic}' is ambiguous. Consider one of the following:")
        for option in e.options:
            print(option)
    except wikipedia.exceptions.HTTPTimeoutError:
        print("The request to Wikipedia timed out. Please try again later.")
    except wikipedia.exceptions.RedirectError:
        print(f"The topic '{topic}' has been redirected.")
    except wikipedia.exceptions.PageError:
        print(f"The topic '{topic}' does not exist on Wikipedia.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
for topic in final_list_topics:
    summary= fetch_wikipedia_summary(topic, sentences=25)
    if summary:
        append_in_csv(summary)
        
```
